```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import and read the csv file
earthquake = pd.read_csv('all_month.csv')
earthquake.dropna(inplace=True)
# here I dropped any NaN values from the data
earthquake
```

```python

plt.scatter(earthquake['mag'], earthquake['depth'])
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.show()
```

```python
corr_mat = earthquake.corr()
corr_mat
```

```python

plt.figure(figsize=(8,8))
sns.heatmap(corr_mat, annot=True)
plt.title('Correlation Plot')
plt.show()
```

```python
plt.bar(earthquake['mag'], earthquake['dmin'])
plt.xlabel('Magnitude', fontweight = "bold")
plt.ylabel('Dmin (Horizontal distance from epicenter to the nearest station (in km))', fontsize=7, fontweight = "bold")
plt.show()
```
